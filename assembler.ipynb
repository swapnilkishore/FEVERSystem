```python
import wikipedia as wki
from textblob import TextBlob as txb
import pandas as pd
import json
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import math
import statistics
from operator import itemgetter
import random
import ast
```

```python
test_claims = []
with open('shared_task_dev.jsonl', 'r') as f:
    for line in f:
        test_claims.append(json.loads(line))
test_claims[9350]
```

```python
#COMBINED SCORE WITH PHRASES
test_evidence = []
with open('accurate_test_output_test_Rake.jsonl', 'r') as f:
    for line in f:
        test_evidence.append(json.loads(line))
test_evidence[:2]
```

```python
test_labels = []
file1 = open("Rake_labels.txt","r") 
for label in file1:
    label = label.replace("\n", "")
    test_labels.append(label)
test_labels
```

```python
counter = 0
with open("predictions_rake.jsonl", "w") as outfile:
    for claim, evidence, label in zip(test_claims[9350:], test_evidence, test_labels):
        #if claim['claim'] == evidence['claim']:
        counter += 1
        output_row = {}
        output_row["label"] = claim['label']
        output_row["predicted_label"] = label
        temp_evidence = []
        for source in evidence['sources']:
            source[1] = int(source[1])
            #print(source)
            temp_evidence.append(source)
        output_row["predicted_evidence"] = temp_evidence
        temp_evidence1 = []
        for source in claim['evidence']:
            temp_evidence1.append(source)
        output_row["evidence"] = temp_evidence1
        json.dump(output_row, outfile)
        outfile.write("\n")
        print(counter)
```
